```python
from gurobipy import *
```

```python
#Problem1 Case1
f = Model("Cashflow Optimization")
c1=f.addVar(name='c1', vtype=GRB.CONTINUOUS)
c2=f.addVar(name='c2', vtype=GRB.CONTINUOUS)
c3=f.addVar(name='c3', vtype=GRB.CONTINUOUS)
c4=f.addVar(name='c4', vtype=GRB.CONTINUOUS)
c5=f.addVar(name='c5', vtype=GRB.CONTINUOUS)
c6=f.addVar(name='c6', vtype=GRB.CONTINUOUS)

cp1=f.addVar(name='cp1', vtype=GRB.CONTINUOUS)
cp2=f.addVar(name='cp2', vtype=GRB.CONTINUOUS)
cp3=f.addVar(name='cp3', vtype=GRB.CONTINUOUS)
cp4=f.addVar(name='cp4', vtype=GRB.CONTINUOUS)
cp5=f.addVar(name='cp5', vtype=GRB.CONTINUOUS)

cl1=f.addVar(name='cl1', vtype=GRB.CONTINUOUS)
cl2=f.addVar(name='cl2', vtype=GRB.CONTINUOUS)
cl3=f.addVar(name='cl3', vtype=GRB.CONTINUOUS)
cl4=f.addVar(name='cl4', vtype=GRB.CONTINUOUS)
cl5=f.addVar(name='cl5', vtype=GRB.CONTINUOUS)

r = 0.03

f.setObjective(c6, GRB.MAXIMIZE)

f.addConstr(-200 -c1 + cp1 + cp2 + cl1 == 0)
f.addConstr( 100 -c2 + c1*1.005 - 1.01 * cp1 + cp3 + cl2 - (1 + r) * cl1 == 0)
f.addConstr(-150 -c3 + c2*1.005 + cp4 + cl3 - (1 + r) * cl2 == 0)
f.addConstr(  50 -c4 + c3*1.005 - 1.05 * cp2 + cp5 + cl4 - (1+r) * cl3 == 0)
f.addConstr(-300 -c5 + c4*1.005 - 1.04 * cp3 + cl5 - (1+r) * cl4 == 0)
f.addConstr( 600 -c6 + c5*1.005 - 1.05 * cp4 - 1.03 * cp5 - (1 + r) * cl5 == 0)

f.addConstr(cl1 <= 100)
f.addConstr(cl2 <= 100)
f.addConstr(cl3 <= 100)
f.addConstr(cl4 <= 100)
f.addConstr(cl5 <= 100)

f.optimize()
print('-----------------')
print('optimal solution is ...')
for i in f.getVars():
    print(i.varname, '=', i.x)
```
